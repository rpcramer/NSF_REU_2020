```python
%%time
dic = dict()
#Process the sentences that contain a skill against other sentences
for doc in docs:
    unique_skills = list()
    
    for tokens in doc:
        if(tokens.i + 1) < len(doc):
            if tokens.is_punct or doc[tokens.i + 1].is_punct:
                continue
            token = tokens.lemma_.lower()
            next_token = doc[tokens.i + 1].lemma_.lower()
            both_token = token + " " + next_token
            if token not in stopwords_list and next_token not in stopwords_list:
                if token not in unique_skills:
                    if token in dic:
                        dic[token] = dic[token] + 1
                    else:
                        dic.update( {token : 1} )
                    unique_skills.append(token)
                if next_token not in unique_skills:
                    if next_token in dic:
                        dic[next_token] = dic[next_token] + 1
                    else:
                        dic.update( {next_token : 1} )
                    unique_skills.append(next_token)
            
                if both_token not in unique_skills:
                    if both_token in dic:
                        dic[both_token] = dic[both_token] + 1
                    else:
                        dic.update( {both_token : 1} )
                    unique_skills.append(both_token)
```
